cr_m és el cratio del entropy i speed_m és el speed del entropy.

a, b són els pesos que se li assignen a cratio i speed respectivament.
a = 1/cr_out, sent cr_out el cratio (real) de la combinació que ha sortit com a output de l’anterior época.
b = 1/speed_out, sent cr_out el speed (real) de la combinació que ha sortit com a output de l’anterior época.

Favor és l’elecció a minimitzar: speed, cratio o balance.
Si volem máxim speed → c = 1
Si volem mínim cratio → c = 1/100
Si volem balance → c = 1/10

Per tant, la funció de pèrdua és: 1 / (a * cr_m + f * b *speed_m).
